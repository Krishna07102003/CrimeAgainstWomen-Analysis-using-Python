Distribution of Crimes

```python
CrimeData_2021 = CrimeData[CrimeData['Year'] == 2021]
incident_types = ['Rape', 'K&A', 'DD', 'AoW', 'AoM', 'DV', 'WT']
total_incidents = CrimeData_2021[incident_types].sum()
print(total_incidents)
```

Pie Chart

Which is the topmost incidents in 2021?

```python
plt.figure(figsize=(8, 8))
plt.pie(total_incidents, labels=incident_types, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Total Incidents by Type (2021)')
plt.axis('equal')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Conclusion : Based on Pie Chart , Domestic Voilence is the topmost incidents happened in 2021

```python
What is the most frequent incidents that increasing rapidly over the years?
```

```python
yearly_crimes = CrimeData.groupby('Year').agg({
    'Rape': 'sum',
    'K&A': 'sum',  # Kidnapping & Abduction
    'DD': 'sum',   # Dowry Deaths
    'AoW': 'sum',  # Assault on Women
    'AoM': 'sum',  # Acid on Women
    'DV': 'sum',   # Domestic Violence
    'WT': 'sum'    # Women Trafficking
}).reset_index()
# Create the line chart
plt.figure(figsize=(15, 8))

# Plot each crime type
plt.plot(yearly_crimes['Year'], yearly_crimes['Rape'], marker='o', label='Rape')
plt.plot(yearly_crimes['Year'], yearly_crimes['K&A'], marker='s', label='Kidnapping & Abduction')
plt.plot(yearly_crimes['Year'], yearly_crimes['DD'], marker='^', label='Dowry Deaths')
plt.plot(yearly_crimes['Year'], yearly_crimes['AoW'], marker='D', label='Assault on Women')
plt.plot(yearly_crimes['Year'], yearly_crimes['AoM'], marker='v', label='Acid on Women')
plt.plot(yearly_crimes['Year'], yearly_crimes['DV'], marker='*', label='Domestic Violence')
plt.plot(yearly_crimes['Year'], yearly_crimes['WT'], marker='p', label='Women Trafficking')

plt.title('Trend of Crimes Against Women (2001-2021)', fontsize=14, pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Cases', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show the plot
plt.show()

# Print the yearly totals
print("\nYearly Totals:")
print(yearly_crimes)
```

Conclusion: Domestic Violence is the most number of increasing incidents over the years

To showing the number of rape cases by states

```python
# Remove duplicates if any
df = CrimeData.drop_duplicates(subset=['State'])

# Sort values in descending order
df_sorted = CrimeData.sort_values('Rape', ascending=True)

# Create the horizontal bar plot
plt.figure(figsize=(12, 12))

# Create bars
plt.barh(df_sorted['State'], df_sorted['Rape'], color='crimson')

# Customize the chart
plt.title('Number of Rape Cases by State', pad=20, fontsize=14)
plt.xlabel('Number of Cases', fontsize=12)
plt.ylabel('States', fontsize=12)


# Add grid lines
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

# Print the statistics
print("\nRape Cases by State:")
print(df_sorted[['State', 'Rape']].to_string(index=False))
```
